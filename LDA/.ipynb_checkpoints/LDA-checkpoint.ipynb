```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from itertools import product
from functools import reduce
%matplotlib inline 
```

```python
iris_raw = datasets.load_iris()
iris = zip(iris_raw["target"], iris_raw["data"])
iris = list(zip(*[[t,d[0:2]] for t,d in iris if t in [0,1]]))
Y = list(iris[0])
X = list(iris[1])
```

```python
class LDA:
    
    def __init__(self, X, Y):
        self.X = X
        self.Y = Y
    
    def train(self):
        prevalance = self._get_prevalance(1)
        mean_0 = self._get_mean(0)
        mean_1 = self._get_mean(1)
        sigma = self._get_sigma()
        return prevalance, mean_0, mean_1, sigma 
    
    def test(self,input):
        pass
    
    def _get_prevalance(self,label):
        X = self.X
        Y = self.Y
        m = len(Y)
        Y_conditioned_sum = sum([1 for y in Y if y == label])
        prevalance = np.asarray(Y_conditioned_sum / m)
        return prevalance 
    
    def _get_mean(self, label):
        X = self.X
        Y = self.Y
        x_conditioned_sum = sum([x for x,y in zip(X,Y) if y == label])
        Y_conditioned_sum = sum([1 for y in Y if y == label])
        mean = np.asarray(x_conditioned_sum / Y_conditioned_sum)
        return mean
    
    def _get_sigma(self):
        X = self.X
        Y = self.Y
        m = len(Y)
        mean_0 = self._get_mean(0)
        mean_1 = self._get_mean(1)
        sigma = np.sum([np.outer((x - mean_0),(x - mean_0)) if y == 0 
                        else np.outer((x - mean_1),(x - mean_1) )
                        for x,y in zip(X,Y)], axis = 0)/m
        return sigma
```

```python
def multi_gausss(x , cov_mtx, mean_vec):
    denominator = (2 * np.pi) * np.linalg.det(cov_mtx)**0.5
    numerator = - 0.5 * reduce(lambda A,B: np.matmul(A,B),[(x - mean_vec),
                                                          np.linalg.inv(cov_mtx),
                                                          (x - mean_vec)])
    numerator = np.exp(numerator)
    prob_x = numerator / denominator
    return prob_x


def plot_contour(cov_mtx, mean_vec):
    print(mean_vec)
    x = np.linspace(4,7)
    y = np.linspace(2,5)
    x,y = np.meshgrid(x,y)
    x_y = list(zip(*[x.ravel(), y.ravel()]))
    z = np.array([multi_gausss(np.asarray(feature),cov_mtx, mean_vec) 
                  for feature in x_y] )
    plt.contour(x,y,z.reshape(x.shape), )
```

```python
model_1 = LDA(X,Y)
prevalance, mean_0, mean_1, sigma = model_1.train()
plot_contour(sigma,mean_0)
plot_contour(sigma,mean_1)
marker_dict = {0:"s", 1:"d"}
for gt,v in marker_dict.items():
    X_1 = [x[0] for x,y in zip(X,Y) if y == gt]
    X_2 = [x[1] for x,y in zip(X,Y) if y == gt]
    plt.scatter(X_1, X_2, marker = v)
    plt.show()
```
